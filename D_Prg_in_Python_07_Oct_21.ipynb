# Merging for variables

```python
# Merging means combining two tables based on a common column or key

# Inner merge (or Inner join)
# Intersection i.e. rows common to both the tables

# Outer merge (or Outer join)
# Union i.e. all the rows

# Left merge or Left join
# All the records from the left table and matched records from the right table

# Right merge or Right join
# All records from the right table and matched records from the left table
```

```python
# Importing the left and right table

left = pd.read_csv('R1.csv')
right = pd.read_csv('R2.csv')
```

```python
left
```

```python
right
```

```python
# Merging based on the common key ID

# Inner join (i.e intersection)

# By default merge will perform inner join
inner_join = pd.merge(left,right, on='ID')
```

```python
inner_join
```

```python
inner_join1 = pd.merge(left,right, on='ID',how='inner')
```

```python
inner_join1
```

```python
# Outer join

outer_join = pd.merge(left,right, on='ID',how='outer')
```

```python
outer_join
```

```python
# Left join

left_join = pd.merge(left,right, on='ID',how='left')
```

```python
left_join
```

```python
# Right join

right_join = pd.merge(left,right, on='ID',how='right')
```

```python
right_join
```

```python
# Importing

WHO = pd.read_csv('WHO_data.csv')
```

```python
WHO.head()
```

```python
# For Categorical columns

WHO.Region.mode()  # category that has the maximum frequency
```

```python
# Frequency for all the categories in the 'Region' column

WHO.Region.value_counts()
```

```python
# To get the unique categories in the 'Region' column

WHO.Region.unique()
```

```python
WHO.head()
```

```python
# Grouping the variables

#1. Single variable, Single statistic

# Sum of all the Numeric columns for each category in the 'Region' column

WHO.groupby(['Region']).sum()
```

```python
WHO.groupby(['Region']).sum().reset_index()
```

```python
#2. Calculating the Grand Total for each Numeric column

WHO.groupby(['Region']).sum().sum()
```

```python
WHO.groupby(['Region']).sum().sum().reset_index()
```

```python
# Total sum

WHO.groupby(['Region']).sum().sum().sum()
```

```python
# 3. Single variable, Multiple Statistics

WHO.groupby('Region').agg({'Population':[min,max, np.mean, np.median,np.std]})
```

```python
# Multiple Variables and Multiple Statistics for a Single Numeric column

WHO.groupby(['Country','Region']).agg({'Population':[min,max, np.mean, np.median]})
```

```python
# 4. Multiple Variables, Multiple Statistics

# values ---> Numeric columns
# index ----> Categorical column
# aggfunc---> Specific statistics for specific Numeric columns

pd.pivot_table(WHO, values=['Population','Under15'],index=['Region'],aggfunc={'Population':[np.mean,min],'Under15':[sum,max]}).reset_index()
```

```python
WHO.head()
```

```python
# Reshaping the data

# Changing the shape or format of the data

WHO11 = WHO[['Country','Region','Population']]
```

```python
WHO11.head()
```

```python
# Longitudinal format (or Tall format)  and Latitudinal format (or Wide format)
```

```python
WHO11
```

```python
# Using pivot() reshaping the data from Longitudinal format to Latitudinal format

WHO12 = WHO11.pivot(index='Region',columns='Country',values='Population').reset_index()
```

```python
WHO12
```

```python
WHO13 = WHO12.fillna(value=0)
```

```python
WHO13
```

```python
# Using melt() reshaping the data from Longitudinal to Latitudinal format

# id_vars ---> column to use as variable
# var_name ---> column we want to unpivot

WHO13.melt(id_vars=['Region'],var_name = 'Country')
```
